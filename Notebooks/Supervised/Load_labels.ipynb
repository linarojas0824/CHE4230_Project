```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from umap import UMAP
import matplotlib.pyplot as plt
```

Use DR and Clustering technique without splitting the data

```python
import pandas as pd
from CHE4230project_LLGModel.KMeansClustering import KMeansClustering
from CHE4230project_LLGModel.DR_UMAP import DR_UMAP
UMAP_normal = DR_UMAP(n_components=2)
df_UMAP_normal = UMAP_normal.dt(data)
```

Clustering for all the data

```python

# Instantiate KMeans object
kmeans = KMeansClustering (num_clusters=5)
# Fit and predict clusters
y_kmeans = kmeans.fit_predict(df_UMAP_normal)
```

Labels and plot to understand correlation

```python
labels_ = labels.to_numpy()
labels_
unique_labels_test = np.unique(labels_[:,1])
unique_labels_y = np.unique(y_kmeans)
```

clustering K-Means Labels

```python
for label in unique_labels_y:
    plt.scatter(df_UMAP_normal[y_kmeans == label, 0], df_UMAP_normal[y_kmeans == label, 1], label=f'Cluster {label}')

plt.legend()
plt.show()
```

Clustering Test Labels

```python
for label in unique_labels_test:
    plt.scatter(df_UMAP_normal[labels_[:,1] == label, 0], df_UMAP_normal[labels_[:,1] == label, 1], label=f'Cluster {label}')

plt.legend()
plt.show()
```
